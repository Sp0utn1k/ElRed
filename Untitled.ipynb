```python
testset = xy_zones[0]
x_min = testset.x.min(); x_max = testset.x.max()
y_min = testset.y.min(); y_max = testset.y.max()
r = np.sqrt(testset.a*testset.b)
g = np.median(r) / 2
Nx = int((x_max - x_min) // g) + 1
Ny = int((y_max - y_min) // g) + 1
counts = np.zeros((Nx, Ny))

ix = ((testset.x - x_min) // g).astype(int)
iy = ((testset.y - y_min) // g).astype(int)

for i,j in zip(ix, iy):
    counts[i,j] += 1
    
counts.shape
```

```python
from scipy.ndimage import gaussian_filter
sigma = 3
blurred = gaussian_filter(counts, sigma=sigma)
thresh = 10 / (2*np.pi * sigma**2)
mask = blurred > thresh
```

```python
from numba import njit

def find_groups(mask, counts):
    visited = np.zeros(mask.shape, dtype=bool)
    return find_groups_jit(mask, visited, counts)

@njit(cache=True)
def find_groups_jit(mask, visited, counts):
    Ni, Nj = mask.shape
    
    groups = []

    directions = [
        (-1,0), (1,0), (0,-1), (0,1)
    ]

    for i in range(Ni):
        for j in range(Nj):
            if mask[i,j] and not visited[i,j]:
                current_group = []
                q = [(i,j)]
                visited[i,j] = True
                
                while len(q):
                    curr_i, curr_j = q[0]
                    q = q[1:]
                    current_group.append([curr_i, curr_j])
                    for di, dj in directions:
                        i2 = curr_i + di
                        j2 = curr_j + dj
                        if 0 <= i2 < Ni and 0 <= j2 < Nj:
                            if mask[i2,j2] and not visited[i2,j2]:
                                visited[i2,j2] = True
                                q.append((i2,j2))
                            
                groups.append(np.array(current_group))
                
                
    return groups
```

```python
groups = find_groups(mask, counts)
groups = sorted(groups, key=len, reverse=True)
len(groups)
```
